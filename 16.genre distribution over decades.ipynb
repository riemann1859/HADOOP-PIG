```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
add="C:/Users/YUNUS/DESKTOP/genre.txt"

d={}
d['genre']=[]
d['decade']=[]
d['perc']=[]
with open(add,mode='r') as file:
    for line in file:
        a=line.strip()[1:-1].split(',')
        d['genre'].append(a[0])
        d['decade'].append(a[1])
        d['perc'].append(float(a[2]))

df=pd.DataFrame(d)
pivot=pd.pivot_table(df, values='perc', columns='decade',index='genre',fill_value=0)
pivot
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(pivot, annot=True, annot_kws={"size": 10})
plt.show()
```
